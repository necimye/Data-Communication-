```python
# Imports
from scipy.fft import fft, ifft, fftfreq
import scipy.signal as sig
import numpy as np
import matplotlib.pyplot as plt
```

```python
start = 0
end = 3
sample_points = 300
t = np.linspace(start, end, sample_points, endpoint=True)
cos_t = 19 * np.cos(2 * np.pi * 5 *  t)
plt.plot(t, cos_t)
plt.xlabel("t")
plt.ylabel("cos(2 * pi * 5 * t)")
plt.title("Cosine wave with frequencey of 5")
```

```python
# Performing fourier transform

f_cos_t = fft(cos_t)
f_s = fftfreq(sample_points, (end-start)/sample_points)
plt.plot(f_s, f_cos_t)
plt.grid()
plt.xlabel("f")
plt.ylabel("F[cos(t)]")
plt.title("Fourier transform of cosine")
plt.show()
```

```python
# Performing the inverse fourier transform
inverse_cosine_fft = ifft(f_cos_t)
plt.plot(t, inverse_cosine_fft)
plt.xlabel("t")
plt.ylabel("F^(-1)[F[cos(t)]]")
plt.title("Inverse fourier transform to obtain original signal")
plt.show()
```

```python
# Carrying out the entire operation for another signal
square_t = 19 * sig.square(2 * np.pi * 5 * t)
plt.plot(t, square_t)
plt.xlabel("t")
plt.title("Square wave of frequency 5")
plt.show()

# Fourier transform
f_sq = fft(square_t)
f_s = fftfreq(sample_points, (end-start)/sample_points)
plt.plot(f_s, f_sq)
plt.grid()
plt.xlabel("f")
plt.title("Fourier transform of square wave")
plt.show()

# Inverse fourier transform
inverse_square_fft = ifft(f_sq)
plt.plot(t, inverse_square_fft)
plt.xlabel("t")
plt.title("Inverse fourier transform to obtain original signal")
plt.show()
```
